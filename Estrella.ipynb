```python
import numpy as np
import matplotlib.pyplot as plt

def plot_star():
    angles = np.linspace(0, 2*np.pi, 6, endpoint=False)
    x = np.cos(angles)
    y = np.sin(angles)
    
    plt.figure(figsize=(6, 6))
    plt.plot(x, y, marker='*', markersize=10, color='blue')
    plt.axis('equal')
    plt.axis('off')
    plt.title('Estrella')
    plt.show()

plot_star()
```
